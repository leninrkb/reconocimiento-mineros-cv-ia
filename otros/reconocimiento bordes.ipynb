```python
import cv2
```

```python
DATADIR = '/home/lenin/Documents/datasets/colores'
```

```python
path = DATADIR + '/testimg' + '/92.jpg'
img = cv2.imread(path)

# Convertir a escala de grises
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Aplicar suavizado Gaussiano para reducir el ruido
gray = cv2.GaussianBlur(gray, (3,3), 0)

# Aplicar detecci√≥n de bordes Canny
edges = cv2.Canny(gray, 50, 150)

# Encontrar contornos en la imagen
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibujar contornos en la imagen
cv2.drawContours(img, contours, -1, (0, 255, 0), 2)

# Mostrar la imagen con los contornos
cv2.imshow("Resultado", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
